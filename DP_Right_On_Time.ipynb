# Intro

A few friends and I are playing a game at 6pm. I know they often arrive late, and I really don't want to be late.

I actually looked at historical data, and found out my friends arrival time can be described as random variable that follows a normal distribution of parameter mean = 'appointment time' amd std_dev = 10 min. 

I need to set the appointment so that Im at least 99% confident all my friends will be there by 6pm. Can you figure out the latest appointment time that would check that criteria? Thanks!!

```python
game_time = 18 * 60 # 6pm in number of minutes from midnight
n_friends = 4
std_dev_arrival_time_min = 10
# ... and that's all!
# right_appointment_time = ?
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def simulate_one_appointment(n_friends, std_to_appointment):  
  return max([np.random.normal(0, std_to_appointment) for i in range(n_friends)])
```

```python
simulate_one_appointment(n_friends, std_dev_arrival_time_min)
```

```python
simulations = [simulate_one_appointment(n_friends, std_dev_arrival_time_min) for i in range(1000000)]
```

```python
time_late = math.ceil(np.percentile(simulations, 99))
```

```python
print(f'99% of people are late at most {time_late} minutes.')
```

```python
meeting_time = game_time - time_late
```

```python
print(f'Meeting time should be set to {math.floor(meeting_time/60)}:{meeting_time%60}.')
```
